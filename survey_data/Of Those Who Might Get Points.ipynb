```python
import pandas as pd
```

```python
df = pd.read_json('results-sanitzed.json')

df.head()
```

```python
mask = df['Global_Leaderboard_Participation'].isin({'Yes, I might get points', 'Yes, I will likely get points'})

df[mask]['Global_Leaderboard_Participation'].value_counts()
```

```python
all_langs = []

for langs in df[mask]['Languages']:
    all_langs.extend(langs)

from collections import Counter

counts = Counter(all_langs).most_common()
```

```python
total = sum([c for lang, c in counts])

for lang, c in counts:
    print('{:30}: {:.3f}%'.format(lang, c / total * 100))
```
